#### Correlation Between Categorial Features

```python
def cramers_v(var1,var2):
    crosstab = pd.crosstab(var1, var2)  # Cross table
    chi2 = chi2_contingency(crosstab)[0]  # Chi-squared statistic
    n = crosstab.values.sum()  # Total number of observations
    r, k = crosstab.shape  # Dimensions of the contingency table
    return np.sqrt(chi2 / (n * (min(k - 1, r - 1))))
    
```

#### Building Of Matrix

```python
rows = []
for var1 in x_train.columns:
    col = []
    for var2 in x_train.columns:
        cramers = cramers_v(x_train[var1], x_train[var2])
        col.append(round(cramers, 2))
    rows.append(col)

# Convert the result to a DataFrame
cramers_result = np.array(rows)
df3 = pd.DataFrame(cramers_result, columns=x_train.columns, index=x_train.columns)
```

```python
def cramers_correlation(dataset, threshold):
    col_corr = set()
    for i in range(len(dataset.columns)):
        for j in range(i):
            var1, var2 = dataset.columns[i], dataset.columns[j]
            crosstab = pd.crosstab(dataset[var1], dataset[var2])
            chi2 = chi2_contingency(crosstab)[0]
            n = crosstab.values.sum()
            r, k = crosstab.shape
            cramers_v = np.sqrt(chi2 / (n * (min(k - 1, r - 1))))
            if cramers_v < threshold:
                col_corr.add(var1)
    return col_corr

# Example Usage:
# Set threshold based on domain expertise
corr1 = cramers_correlation(x_train, 0.05)
print("low correlated features:", corr1)
```

### Create an Evaluate Function to give all metrics after model training

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Define the regression models
models = {
    "XGBoost Regressor": xgb.XGBRegressor(),
    "Support Vector Regressor": SVR(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Decision Tree Regressor": DecisionTreeRegressor(),
    "KNeighbors Regressor": KNeighborsRegressor(),
    "Linear Regression": LinearRegression(),
    "CatBoost Regressor": CatBoostRegressor(verbose=0),  # Suppress CatBoost's verbose output
    "Ridge Regressor": Ridge(),
    "Lasso Regressor": Lasso(),
    "AdaBoost Regressor": AdaBoostRegressor()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train, y_train)

    # Make Predictions
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # Calculate metrics for the training set
    train_mae = mean_absolute_error(y_train, y_train_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    train_rmse = np.sqrt(train_mse)
    train_r2 = r2_score(y_train, y_train_pred)

    # Calculate metrics for the test set
    test_mae = mean_absolute_error(y_test, y_test_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    test_rmse = np.sqrt(test_mse)
    test_r2 = r2_score(y_test, y_test_pred)

    # Print results
    print(list(models.keys())[i])
    print("-----------Metrics for Train Data------------")
    print(f"Mean Absolute Error (Train): {train_mae:.4f}")
    print(f"Mean Squared Error (Train): {train_mse:.4f}")
    print(f"Root Mean Squared Error (Train): {train_rmse:.4f}")
    print(f"R^2 Score (Train): {train_r2:.4f}")

    print("-----------Metrics for Test Data------------")
    print(f"Mean Absolute Error (Test): {test_mae:.4f}")
    print(f"Mean Squared Error (Test): {test_mse:.4f}")
    print(f"Root Mean Squared Error (Test): {test_rmse:.4f}")
    print(f"R^2 Score (Test): {test_r2:.4f}")
    print('='*35)
    print('\n')
```
